```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv(r"C:\Users\bollejayanthsriteja\OneDrive\Desktop\code\user_engagement_dataset.csv")
```

```python
# Display basic information
print("Dataset Info:")
print(data.info())
```

```python
# Display first few rows
print("\nSample Data:")
print(data.head())
```

```python
# Data Cleaning and Preprocessing
data['Application_Date'] = pd.to_datetime(data['Application_Date'])  # Convert date
print("\nMissing Values:")
print(data.isnull().sum())
```

```python

# --- ANALYSIS AND VISUALIZATION ---

# 1. Most popular domains
popular_domains = data['Domain'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_domains.index, y=popular_domains.values, palette='viridis')
plt.title('Most Popular Internship Domains')
plt.xlabel('Domain')
plt.ylabel('Number of Applications')
plt.xticks(rotation=45)
plt.show()
```

```python

# 2. Participation status distribution
participation_counts = data['Participation'].value_counts()
plt.figure(figsize=(6, 6))
participation_counts.plot.pie(autopct='%1.1f%%', colors=['#4CAF50', '#FF5722'], startangle=90)
plt.title('Participation Status')
plt.ylabel('')
plt.show()
```

```python
# 3. Applications over time
applications_by_date = data.groupby('Application_Date').size()
plt.figure(figsize=(12, 6))
applications_by_date.plot(color='teal')
plt.title('Applications Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Applications')
plt.show()
```

```python
# 4. University-wise applications
university_applications = data['University'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=university_applications.index, y=university_applications.values, palette='coolwarm')
plt.title('Applications by University')
plt.xlabel('University')
plt.ylabel('Number of Applications')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Average clicks by domain
average_clicks = data.groupby('Domain')['Clicks_on_Internship'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=average_clicks.index, y=average_clicks.values, palette='rocket')
plt.title('Average Clicks by Domain')
plt.xlabel('Domain')
plt.ylabel('Average Clicks')
plt.xticks(rotation=45)
plt.show()
```
